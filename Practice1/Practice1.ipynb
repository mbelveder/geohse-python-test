### Переменные

- Нужны для хранения информации
- Бывают разные типов:
    - `int` или `integer`: для хранения целых чисел
    - `bool`: для хранения `True` или `False`
    - `float`: для хранения нецелых чисел
    - `string`: для хранения строк (символов или слов)
    - Других типов
- Хранятся в памяти `python` если инициализированы (выполнен код где определяется переменная)

````python
a = 3
b = True
c = 3.14
d = 'Доктор это вам'
````

````python
print('Значение переменной a:', a, 'Тип переменной a:', type(a))
print('Значение переменной b:', b, 'Тип переменной b:', type(b))
print('Значение переменной c:', c, 'Тип переменной c:', type(c))
print('Значение переменной d:', d, 'Тип переменной d:', type(d))
````

### Операции над переменными

- Основные операторы:
    - `+`
    - `-`
    - `/`
    - `*`
- Операторы сравнения (как результат возвращают тип `bool`):
    -  `>` - больше, `>=` - больше или равно
    -  `<` - меньше, `<=` - меньше или равно
    -  `==` - проверка равенства (возрващает два если значение одинаково)
    -  `!=` - не равно
- Логические операторы:
    - `and`
    - `or`
    - `not`
- **Дополнительные операторы** (забыл рассказать на лекции, но сейчас будут нам нужны):
    - `x**n` - `x` в степени `n`
    - `//` - деление нацело
    - `%` - остаток от деления

````python
print('10 + 2 =', 10 + 2)
print('10 - 2 =', 10 - 2)
print('10/2 =', 10/2) # обратите внимание что результат уже не integer
print('10*2 =', 10*2)
print('10 > 2 =', 10 > 2)
print('10 < 2 =', 10 < 2)
print('10 == 2 =', 10 == 2)
print('10 != 2 =', 10 != 2)
print('True and False:', True and False)
print('True or False:', True or False)
print('not True:', not True)
print('2**3 =', 2**3)
print('9 // 4 =', 9 // 4)
print('9 % 4 =', 9 % 4)
````

### Условные операторы:

- оператор `if` позволяет выполнять блок кода при выполнении условия
- следующая конструкция:
    - `if` + условие + :
    - (отступ) Блок кода
    - Блок кода вне отступа выполняется без условий
- `else`: блок кода, который будет выполняться если условие не выполнилось

````python
number = 5 # можно поменять число

if number//2:
    print('Число четное')
else:
    print('Число нечетное')
````
